1.B) Usar um pacote pronto do python para aplicar a regressão linear na base de dados. Fazer os

```python
data = {
    'Base 1': base_1,
    'Base 2': base_2
}

df = pd.DataFrame(data)

corr, _ = pearsonr(df['Base 1'], df['Base 2'])
print('Coeficiente de correlação:', corr)

x_15= df['Base 1'].values.reshape(-1, 1)
y_15= df['Base 2'].values

model = LinearRegression()
model.fit(x_15, y_15)

a_15 = model.coef_[0]
b_15 = model.intercept_

print('Coeficiente angular (a):', a_15)
print('Coeficiente linear (b):', b_15)

plt.scatter(x_15, y_15)
plt.plot(x_15, a_15 * x_15 + b_15, color='red', label="Linha de Regressão")
plt.xlabel('Base 1')
plt.ylabel('Base 2')
plt.title('Regressão Linear')
plt.show()
```

1.C) - Encontrar os coeficiêntes a e b, usando as fórmulas dadas em aula programando suas próprias
funções. Fazer os gráficos para a base de dados e para a regressão linear.

```python
def correlacao(x, y):
    n = len(x)
    soma_x = sum(x)
    soma_y = sum(y)
    soma_xy = sum([x[i] * y[i] for i in range(n)])
    soma_x_quadrado = sum([x[i] ** 2 for i in range(n)])
    soma_y_quadrado = sum([y[i] ** 2 for i in range(n)])

    numerador = n * soma_xy - soma_x * soma_y
    denominador = ((n * soma_x_quadrado - soma_x ** 2) * (n * soma_y_quadrado - soma_y ** 2)) ** 0.5

    return numerador / denominador

def regressao_linear(x, y):
    n = len(x)
    soma_x = sum(x)
    soma_y = sum(y)
    soma_xy = sum([x[i] * y[i] for i in range(n)])
    soma_x_quadrado = sum([x[i] ** 2 for i in range(n)])

    a = (n * soma_xy - soma_x * soma_y) / (n * soma_x_quadrado - soma_x ** 2)
    b = (soma_y - a * soma_x) / n

    return a, b

correlacao_15 = correlacao(base_1, base_2)
print('Coeficiente de correlação:', correlacao_15)
coef_angular_15, coef_linear_15 = regressao_linear(base_1, base_2)
print('Coeficiente angular (a):', coef_angular_15)
print('Coeficiente linear (b):', coef_linear_15)
plt.scatter(x_15, y_15)
plt.plot(base_1, [coef_angular_15 * x + coef_linear_15 for x in base_1], color='green', label="Linha de Regressão")
plt.title('Regressão Linear')
plt.xlabel('Base 1')
plt.ylabel('Base 2')
plt.legend()
plt.show()
```
