Exercicio 5.37

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Dados dos irmãos e irmãs
irmaos = np.array([71, 68, 66, 67, 70, 71, 70, 73, 72, 65, 66, 70])
irmas = np.array([69, 64, 65, 63, 66, 65, 64, 66, 59, 62, 66, 64])

# (a) Calcular correlação e equação da reta de regressão
correlation = np.corrcoef(irmaos, irmas)[0, 1]
slope, intercept, _, _, _ = stats.linregress(irmaos, irmas)

# Exibir correlação e equação da reta
print(f"Correlação (r): {correlation:.3f}")
print(f"Equação da reta de regressão: y = {slope:.3f}x + {intercept:.3f}")

# (b) Previsão da altura da irmã Tonya (com base na altura de Damien de 70 polegadas)
altura_damien = 70
predicted_altura_tonya = slope * altura_damien + intercept
print(f"Previsão da altura de Tonya (irmã de Damien): {predicted_altura_tonya:.2f} polegadas")

# (a) Diagrama de dispersão e reta de regressão
plt.scatter(irmaos, irmas, color='blue', label='Dados reais')
plt.plot(irmaos, slope * irmaos + intercept, color='red', label='Reta de regressão')

# Títulos e rótulos
plt.title('Altura de Irmãos vs Irmãs')
plt.xlabel('Altura dos Irmãos (polegadas)')
plt.ylabel('Altura das Irmãs (polegadas)')
plt.legend()
plt.grid(True)
plt.show()
```

Exercicio 5.38

```python
nitratos = np.array([50, 100, 200, 400, 800, 1200, 1600, 2000])
absorbancia = np.array([7.0, 7.5, 12.8, 24.0, 47.0, 93.0, 138.0, 183.0])

# (a) Calcular correlação e equação da reta de regressão
correlation_nitrato_absorbancia = np.corrcoef(nitratos, absorbancia)[0, 1]
slope_nitrato, intercept_nitrato, _, _, _ = stats.linregress(nitratos, absorbancia)

# Previsão do nível de nitrato com absorbância de 40
absorbancia_40 = 40
predicted_nitrato = (absorbancia_40 - intercept_nitrato) / slope_nitrato

# Exibir correlação, equação da reta e previsão
print(f"Correlação (r): {correlation_nitrato_absorbancia:.3f}")
print(f"Equação da reta de regressão: y = {slope_nitrato:.3f}x + {intercept_nitrato:.3f}")
print(f"Previsão do nível de nitrato com absorbância de 40: {predicted_nitrato:.2f} ppm")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Criando um DataFrame a partir dos dados (ajuste conforme sua estrutura real)
data = {'Ano': [1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018],
        'Temperatura': [14.25, 14.37, 14.23, 14.42, 14.56, 14.34, 14.33, 14.47, 14.52, 14.54, 14.49, 14.57, 14.54, 14.52, 14.45, 14.55, 14.63, 14.48, 14.54, 14.58, 14.64, 14.83, 14.90, 14.81, 14.73]}
df = pd.DataFrame(data)

# Separando as variáveis independentes (ano) e dependentes (temperatura)
X = df['Ano'].values.reshape(-1, 1)
y = df['Temperatura'].values

# Criando e treinando o modelo de regressão linear
model = LinearRegression()
model.fit(X, y)

# Fazendo a predição para o ano de 2050
ano_2050 = 2050
predicao_2050 = model.predict([[ano_2050]])

# Coeficientes da reta
coeficiente_angular = model.coef_[0]
coeficiente_linear = model.intercept_

# Plotando o gráfico de dispersão e a reta de regressão
plt.scatter(X, y, label='Dados')
plt.plot(X, model.predict(X), color='red', label='Reta de Regressão')
plt.xlabel('Ano')
plt.ylabel('Anomalia de Temperatura')
plt.title('Anomalia de Temperatura Global vs. Ano')
plt.legend()
plt.show()

# Imprimindo os resultados
print("Equação da reta: y =", coeficiente_angular, "x +", coeficiente_linear)
print("Predição para 2050:", predicao_2050)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parte 1: PERDAS

# Simulação de dados (exemplo) para aversão neural e aversão comportamental
x = np.random.normal(50, 10, 50)  # aversão neural
y = 2 * x + np.random.normal(0, 5, 50)  # aversão comportamental com um pouco de ruído

# (a) Diagrama de dispersão
plt.scatter(x, y)
plt.title("Diagrama de dispersão: Aversão neural vs Aversão comportamental")
plt.xlabel("Aversão Neural")
plt.ylabel("Aversão Comportamental")
plt.show()

# (b) Reta de regressão de mínimos quadrados
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
print(f"Equação da reta de regressão: y = {intercept:.2f} + {slope:.2f}x")

# Adicionando a reta ao diagrama
plt.scatter(x, y)
plt.plot(x, intercept + slope * x, 'r', label=f"y = {intercept:.2f} + {slope:.2f}x")
plt.legend()
plt.show()

# (c) Adicionando um valor atípico
x_outlier = np.append(x, 100)
y_outlier = np.append(y, 50)

plt.scatter(x_outlier, y_outlier)
plt.title("Diagrama de dispersão com valor atípico")
plt.xlabel("Aversão Neural")
plt.ylabel("Aversão Comportamental")

# Recalculando a regressão com o valor atípico
slope_outlier, intercept_outlier, r_value_outlier, _, _ = stats.linregress(x_outlier, y_outlier)
plt.plot(x_outlier, intercept_outlier + slope_outlier * x_outlier, 'r', label=f"y = {intercept_outlier:.2f} + {slope_outlier:.2f}x")
plt.legend()
plt.show()

# Comparando correlações
print(f"Correlação sem valor atípico: {r_value:.2f}")
print(f"Correlação com valor atípico: {r_value_outlier:.2f}")

# Parte 2: MERGULHOS DE PINGUINS

# (a) Inclinação da reta de regressão já foi dada: 0,0138
# (b) Calculando o tempo de mergulho para profundidade de 200 metros
P = 200
DM = 2.69 + 0.0138 * P
print(f"Duração do mergulho para P = 200 metros: {DM:.2f} minutos")

# (c) Gerando a reta de regressão para P = 40 e P = 300
P_values = np.array([40, 300])
DM_values = 2.69 + 0.0138 * P_values

plt.plot(P_values, DM_values, 'o-', label="Reta de Regressão")
plt.title("Reta de regressão: Duração do mergulho vs Profundidade")
plt.xlabel("Profundidade (metros)")
plt.ylabel("Duração do mergulho (minutos)")
plt.legend()
plt.show()
```
