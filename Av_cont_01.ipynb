1.C) Usando funções programadas pelo estudante (Com exceção das visualizações gráficas).


```python
#Media
Media = sum(n)/len(n)
print(f"Média: {Media}")


#Mediana
def cal_mediana(n):
    n_ordenado = sorted(n)
    meio = len(n_ordenado) // 2
    if len(n_ordenado) % 2 == 0:
        return (n_ordenado[meio - 1] + n_ordenado[meio]) / 2
    else:
        return n_ordenado[meio]
Mediana = cal_mediana(n)
print(f"Mediana: {mediana}")


#Variancia amostral
def cal_var_amostral(n):
    media = Media
    soma_diferencas = sum((x - media) ** 2 for x in n)
    return soma_diferencas / (len(n) - 1)

VarianciaAmostral = cal_var_amostral(n)
print(f"Variância Amostral: {VarianciaAmostral}")


#Variancia populacional
def cal_var_populacional(n):
    media = Media
    soma_diferencas = sum((x - media) ** 2 for x in n)
    return soma_diferencas / len(n)

VarianciaPopulacional = cal_var_populacional(n)
print(f"Variância Populacional: {VarianciaPopulacional}")


#Desvio amostral
def cal_desvio_amostral(n):
    variancia_amostral = cal_var_amostral(n)
    return variancia_amostral ** 0.5

print(f"Desvio Padrão Amostral: {cal_desvio_amostral(n)}")


#Desvio populacional
def cal_desvio_populacional(n):
    variancia_populacional = cal_var_populacional(n)
    return variancia_populacional ** 0.5

print(f"Desvio Padrão Populacional: {cal_desvio_populacional(n)}")

#Coeficiente de variação (amostral ou populacional)
CV_amostral = (desvio_padraoAmostral/Media) *100
CV_populacional = (desvio_padraoPopulacional/Media) *100
print(f"Coeficiente de Variação Amostral: {CV_amostral}%")
print(f"Coeficiente de Variação Populacional: {CV_populacional}%")

#Quartis
def cal_quartil(n, percentil):
    n_ordenado = sorted(n)
    k = (len(n_ordenado) - 1) * percentil / 100
    f = int(k)
    c = k - f
    if f + 1 < len(n_ordenado):
        return n_ordenado[f] + (n_ordenado[f + 1] - n_ordenado[f]) * c
    else:
        return n_ordenado[f]

print(f"Primeiro Quartil (Q1): {cal_quartil(n, 25)}")
print(f"Segundo Quartil (Q2/Mediana): {cal_quartil(n, 50)}")
print(f"Terceiro Quartil (Q3): {cal_quartil(n, 75)}")
```
